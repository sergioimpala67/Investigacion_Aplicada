```python
import numpy
file=open("lys1_charges.xyzr",'r')
file_2=open("lys1_charges.stern",'w')

X=numpy.loadtxt(file)
X[:,3] +=2


for i in range(len(X[:,3])):
    x=X[i,:]
    linea = '{0} {1} {2} {3}\n'.format(x[0],x[1],x[2],x[3])
    file_2.write(linea)
file_2.close()
file.close()



    
```
